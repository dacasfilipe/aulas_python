### Resolução do Exercício 1,
"Meu Nome e Idade", pode ser feita de várias maneiras em Python. Abaixo, apresento um exemplo simples:

```python
# Criar variáveis para nome e idade
nome = "Maria"
idade = 30

# Imprimir a mensagem
print(f"Meu nome é {nome} e eu tenho {idade} anos.")
```

Neste exemplo, usei uma f-string (`f"..."`) para incorporar as variáveis `nome` e `idade` diretamente na string que está sendo impressa. Isso torna o código mais legível e fácil de entender.

Alternativamente, você também poderia usar a forma de concatenação de strings ou a função `format`:

**Concatenação de Strings:**
```python
print("Meu nome é " + nome + " e eu tenho " + str(idade) + " anos.")
```

**Usando `.format()`:**
```python
print("Meu nome é {} e eu tenho {} anos.".format(nome, idade))
```

Qualquer uma dessas abordagens dará o resultado desejado.
