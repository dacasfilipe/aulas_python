### Aula 10: Introdução ao Desenvolvimento Web com Flask

#### Duração
- 1,5 a 2 horas

#### Objetivos
- Introduzir o conceito de desenvolvimento web.
- Ensinar como criar uma aplicação web simples com Flask.

#### Material Necessário
- Computador com Python e Flask instalados
- IDE ou editor de texto para programação
- Projetor para exibir slides e exemplos de código

---

#### Parte 1: Introdução ao Desenvolvimento Web

1. **O que é Desenvolvimento Web?**
    - Front-end vs. Back-end
    - Frameworks web

2. **O que é Flask?**
    - Micro framework para Python

---

#### Parte 2: Criando uma Aplicação Flask

1. **Instalando Flask**
    ```
    pip install Flask
    ```

2. **Hello World em Flask**
    ```python
    from flask import Flask
    app = Flask(__name__)

    @app.route('/')
    def hello_world():
        return 'Hello, World!'
    ```

3. **Rodando a Aplicação**
    ```
    flask run
    ```

---

#### Parte 3: Rotas e Templates

1. **Criando Rotas**
    ```python
    @app.route('/about')
    def about():
        return 'About Page'
    ```

2. **Usando Templates**
    - HTML e Jinja2
    ```python
    from flask import render_template

    @app.route('/')
    def home():
        return render_template('home.html')
    ```
